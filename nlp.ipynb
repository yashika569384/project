```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
text_data = [

"NLP is amazing! Learning models and patterns.",

"Learning NLP is fun and powerful.",

"Deep learning improves NLP applications.",

"Text data requires cleaning and preprocessing.",

"Natural Language Processing is widely used in AI.",

"AI models learn from large amounts of text data.",

"Preprocessing text is an important step in NLP.",

"Machine learning and deep learning are related fields."

]
```

```python
import nltk
nltk.download('punkt')
```

```python


from nltk.tokenize import word_tokenize
import string
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
def preprocess(text):
  text=text.lower()
  text=text.translate(str.maketrans('','',string.punctuation))
  tokens=word_tokenize(text)
  stemmedtoken=[stemmer.stem(word) for word in tokens]
  return " ".join(stemmedtoken)

cleaned_text=[preprocess(text) for text in text_data]


```

```python
print(cleaned_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
bow_matrix=bow.fit_transform(cleaned_text)
print(bow_matrix.shape)
```

```python
bow_array=bow_matrix.toarray()
print(bow_array)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
t=tf.fit_transform(cleaned_text)
print(t.shape)
```

```python
tf.vocabulary_
```

```python
feature_names= tf.get_feature_names_out()
print(feature_names)
```

**bag of word**
counts how many times each word appears in a document.onverts text into numerical form by counting the frequency of each word, but it does not consider the importance or context of words.


**TF-IDF**
measures how important a word is by reducing the weight of common words and giving more weight to rare words.improves this by assigning weights to words based on their frequency in a document and their rarity across all documents, giving higher importance to meaningful words.
