```python

from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
from transformers import AutoModelForSeq2SeqLM
import torch

token = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
nlp = pipeline("sentiment-analysis", model=model, tokenizer=token)
```

```python
input_text = """
Climate change is increasingly becoming one of the most pressing issues of the 21st century.
Global temperatures have risen dramatically over the past century, leading to extreme weather events, rising sea levels, and shifts in ecosystems.
Scientists warn that if greenhouse gas emissions continue at the current rate, we could witness catastrophic consequences for both human and natural systems.

In recent years, hurricanes, floods, wildfires, and droughts have caused widespread destruction across the globe.
For example, the 2023 wildfire season in Europe and North America destroyed millions of acres of forest, displaced thousands of people, and caused severe economic losses.
Similarly, flooding in Southeast Asia led to major infrastructure damage, loss of livelihoods, and severe disruptions to agricultural production.
These events highlight the urgent need for countries to implement effective climate policies.

The impact of climate change is not only environmental but also social and economic.
Vulnerable communities, especially those in developing countries, are disproportionately affected due to limited resources and infrastructure.
Displacement due to natural disasters is increasing, leading to climate refugees who face social, political, and economic challenges in their new locations.
Health problems are also on the rise, including heat-related illnesses, respiratory issues due to air pollution, and the spread of vector-borne diseases.

Renewable energy is seen as a key solution to mitigate climate change.
Solar, wind, and hydropower technologies have advanced significantly, making clean energy more affordable and accessible.
Many countries are investing heavily in renewable energy infrastructure, aiming to reduce reliance on fossil fuels and lower carbon emissions.
However, the transition to renewable energy is not without challenges, including technological limitations, initial investment costs, and political resistance.

Climate adaptation strategies are equally important.
Building resilient infrastructure, creating effective early warning systems, and adopting sustainable agricultural practices can help communities cope with the inevitable changes.
Education and public awareness campaigns play a crucial role in changing behavior and encouraging sustainable lifestyles.
Governments, corporations, and individuals all have a role to play in addressing the crisis.

International cooperation is also critical.
Agreements like the Paris Accord aim to bring countries together to set emission targets and collaborate on climate sol
The private sector is increasingly recognizing the importance of sustainability.
Companies are integrating environmental, social, and governance (ESG) criteria into their operations and investment decisions.
Green finance, carbon trading, and sustainable supply chain management are becoming mainstream.
Such initiatives not only reduce environmental impact but also create economic opportunities and build brand value.

Technological innovations offer new ways to combat climate change.
Advances in carbon capture and storage, electric vehicles, and smart grid systems are helping reduce emissions.
Artificial intelligence and data analytics are being used to optimize energy usage, predict climate patterns, and improve disaster response.
Investment in research and development is crucial to accelerate the deployment of these technologies globally.

Despite the growing awareness and action, significant barriers remain.
Political inertia, economic inequalities, and conflicting interests slow down progress.
Public engagement and grassroots movements have shown that citizens can influence policy, but systemic change requires sustained global effort.
Individuals can also make a difference through lifestyle changes, energy conservation, and supporting sustainable products and services.

In conclusion, climate change is a multifaceted challenge that requires immediate and coordinated action.
From governments to corporations to individuals, everyone has a role in mitigating the impacts and adapting to the changes.
Through renewable energy, technological innovation, international cooperation, and public engagement, it is possible to reduce the risks and build a sustainable future.
Failing to act, however, could have dire consequences for future generations and the planet as a whole.
"""
```

```python
max_tokens = token.model_max_length
def chunk_text(text, max_len=max_tokens):
    words = text.split()
    chunks = []
    current = []
    for w in words:
        current.append(w)
        if len(token(" ".join(current))['input_ids']) > max_len:
            chunks.append(" ".join(current[:-1]))
            current = [w]
    if current:
        chunks.append(" ".join(current))
    return chunks

chunks = chunk_text(input_text)
```

```python
predicti = []
for chunk in chunks:
    pred = nlp(chunk)
    predicti.append(pred)
```

```python
for i, pred in enumerate(predicti):
    print(f"Chunk {i+1}: {pred}")
```
